```
from pydrake.all import MathematicalProgram, IpoptSolver
import numpy as np

prog = MathematicalProgram()

N = 20
v_max = 30
a_max = 20 


x = np.zeros((N, 4), dtype="object")
for i in range(N):
    x[i] = prog.NewContinuousVariables(4, "x_" + str(i))

u = np.zeros((N-1, 2), dtype="object")
for i in range(N-1):
    u[i] = prog.NewContinuousVariables(2, "u_" + str(i))

dt = prog.NewContinuousVariables(1, "dt")

t = N * dt[0]
prog.AddConstraint(dt[0] >= 0)
prog.AddBoundingBoxConstraint(np.zeros((4, )), np.zeros((4, )), x[0, :].flatten())
for i in range(1, N-1):
    prog.AddBoundingBoxConstraint(-v_max * np.ones((2, )), v_max * np.ones((2, )), x[i, 2:].flatten())


x_f = 100
y_f = 20
prog.AddBoundingBoxConstraint(x_f, x_f, x[-1, 0])
prog.AddBoundingBoxConstraint(y_f, y_f, x[-1, 1])



for i in range(1, N-1):
    A = np.zeros((4, 4))
    A[0, 2] = 1
    A[1, 3] = 1
    B = np.zeros((4, 2))
    B[2, 0] = 1
    B[3, 1] = 1
    
    prog.AddLinearConstraint(x[i, :] == (A @ x[i - 1, :] + B @ u[i, :]) * dt[0])

    
    # prog.AddBoundingBoxConstraint(-v_max * np.ones((2, )), v_max * np.ones((2, )), x[i, 2:].flatten())


prog.AddQuadraticCost(t**2)


# prog.SetInitialGuess(dt[0], 0.01)


solver = IpoptSolver()
result = solver.Solve(prog)


# u_mpc = result.GetSolution(u[0]) + self.u_d()[0]
```
